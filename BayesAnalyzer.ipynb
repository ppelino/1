
Exercício sobre o Teorema de Bayes:

Vamos considerar um exemplo prático que envolve testes médicos para uma determinada doença. Suponha que uma determinada doença afete 1% da população. Um teste para diagnosticar essa doença tem uma sensibilidade de 90% (probabilidade de dar positivo quando a doença está presente) e uma especificidade de 95% (probabilidade de dar negativo quando a doença está ausente).

Perguntas:

Qual é a probabilidade de um indivíduo realmente ter a doença se o teste der positivo?
Qual é a probabilidade de um indivíduo não ter a doença se o teste der negativo?

```python
# Dados do exercício
prob_doenca = 0.01
prob_positivo_doenca = 0.90
prob_negativo_sem_doenca = 0.95

# Cálculo de P(Positivo)
prob_positivo = prob_positivo_doenca * prob_doenca + (1 - prob_negativo_sem_doenca) * (1 - prob_doenca)

# Cálculo de P(Doença | Positivo)
prob_doenca_positivo = (prob_positivo_doenca * prob_doenca) / prob_positivo

# Exibição da resposta
print(f"A probabilidade de ter a doença se o teste der positivo é: {prob_doenca_positivo:.2%}")
```

```python
Image(filename='Captura de tela 2024-03-02 163609.png')
```

```python
# Cálculo de P(Negativo)
prob_negativo = (1 - prob_positivo_doenca) * prob_doenca + prob_negativo_sem_doenca * (1 - prob_doenca)

# Cálculo de P(Sem Doença | Negativo)
prob_sem_doenca_negativo = (prob_negativo_sem_doenca * (1 - prob_doenca)) / prob_negativo

# Exibição da resposta
print(f"A probabilidade de não ter a doença se o teste der negativo é: {prob_sem_doenca_negativo:.2%}")
```
